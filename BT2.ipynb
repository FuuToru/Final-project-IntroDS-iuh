> Thực hiện thống kê căn bản

```python
import pandas as pd

# Đọc dữ liệu từ file csv
cols = ['CSVC chung','Thiết bị phòng TH','Phòng LT','Tình trạng kẹt xe','Nhà ăn','Mức độ SV vào TV','Không gian','Tài liệu','Máy tính','Thiết bị khác']
data = pd.read_csv('Data_bai2.csv',skiprows=1,usecols=cols)
df = pd.DataFrame(data)
df
```

* Tính toán ma trận tương quan giữa các cột

```python
print(df.corr())
```

* Thể hiện mối tương bằng biểu đồ nhiệt

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(), annot=True)
plt.show()
```

* Biểu đồ thể hiện các đánh giá của mỗi trường

```python
f, ax = plt.subplots(4, 3)

ax[0, 0].pie(df['CSVC chung'].value_counts(), labels=df['CSVC chung'].value_counts().index, shadow=True, autopct='%1.1f%%', startangle=90)
ax[0, 0].set_title('CSVS chung')
plt.grid(True)

ax[0, 1].pie(df['Thiết bị phòng TH'].value_counts(), labels=df['Thiết bị phòng TH'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[0, 1].set_title('Cơ sở phòng thực hành')
plt.grid(True)

ax[0, 2].pie(df['Phòng LT'].value_counts(), labels=df['Phòng LT'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[0, 2].set_title('Cơ sở phòng lý thuyết')
plt.grid(True)

ax[1, 0].pie(df['Tình trạng kẹt xe'].value_counts(), labels=df['Tình trạng kẹt xe'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[1, 0].set_title('Tình trạng kẹt xe ở trường')
plt.grid(True)

ax[1, 1].pie(df['Nhà ăn'].value_counts(), labels=df['Nhà ăn'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[1, 1].set_title('Cơ sở của nhà ăn')
plt.grid(True)

ax[1, 2].pie(df['Mức độ SV vào TV'].value_counts(), labels=df['Mức độ SV vào TV'].value_counts().index, shadow=True, autopct='%1.1f%%', startangle=90)
ax[1, 2].set_title('Mức độ SV vào TV')
plt.grid(True)

ax[2, 0].pie(df['Không gian'].value_counts(), labels=df['Không gian'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[2, 0].set_title('Không gian TV')
plt.grid(True)

ax[2, 1].pie(df['Tài liệu'].value_counts(), labels=df['Tài liệu'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[2, 1].set_title('Tài liệu ở TV')
plt.grid(True)

ax[2, 2].pie(df['Máy tính'].value_counts(), labels=df['Máy tính'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[2, 2].set_title('Máy tính TV cung cấp')
plt.grid(True)

ax[3, 0].pie(df['Thiết bị khác'].value_counts(), labels=df['Thiết bị khác'].value_counts().index, autopct='%1.1f%%', startangle=90)
ax[3, 0].set_title('Thiết bị khác')
plt.grid(True)

ax[3, 1].axis('off')
ax[3, 2].axis('off')

plt.subplots_adjust(top=2, bottom=-0.8, left=0, right=2, hspace=0,
                    wspace=0.2)

plt.show();
```

```python
import matplotlib.pyplot as plt

# Dữ liệu
labels = ['Tốt', 'Trung bình', 'Tệ','Chưa biết']
col = pd.concat([df['CSVC chung'],df['Thiết bị phòng TH'],df['Phòng LT'],df['Tình trạng kẹt xe'],df['Nhà ăn']])
sizes = col.value_counts()


colors = ['#1f77b4', '#ff7f0e', '#2ca02c','#FFA07A']

# # Tạo biểu đồ 
explore  = (0.1,0,0,0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels,explode=explore, colors=colors,shadow=True, autopct='%1.1f%%', startangle=90)
ax1.set_title('Mức độ đánh giá về CSVC của trường')

# # Chỉnh sửa hình dạng của biểu đồ tròn thành hình tròn
ax1.axis('equal')

# # Chú thích
legend = ax1.legend(labels, title="Đánh giá", loc="center right",bbox_to_anchor=(0.7, 0, 0.5, 1))
plt.setp(legend.get_title(), fontsize='large')

# # Hiển thị biểu đồ tròn
plt.show()

```
